```python
import streamlit as st
import glob
from st_speckmol import speck_plot
import st_speckmol
import numpy as np
import pandas as pd


coordinates2 = np.array(np.genfromtxt('data.csv', dtype=float, delimiter=","))
print(coordinates2)

def transform_to_cube(coordinates, desired_size=None):
    # Step 1: Calculate the center of mass
    center_of_mass = np.mean(coordinates, axis=0)

    # Step 2: Translate the molecule to the origin
    translated_coordinates = coordinates - center_of_mass

    # Step 3: Find the maximum coordinate range
    max_range = np.max(np.abs(translated_coordinates))

    # Step 4: Scale the coordinates
    scaled_coordinates = translated_coordinates / max_range

    # Step 5: Rescale to desired size (optional)
    if desired_size is not None:
        scaled_coordinates *= desired_size

    return scaled_coordinates

# Example usage
coordinates = np.array([
    [1.000, 1.983, 8.300],
    [2.000, 1.983, 8.000],
    [3.000, 1.983, 8.000],
    [4.000, 1.983, 8.300],
    [5.000, 1.983, 8.500],
    [6.000, 1.983, 8.800],
    [7.000, 1.983, 8.900],
    [8.000, 1.983, 9.000],
    [9.000, 1.983, 9.100],
    [10.000, 1.983, 9.200],
    [11.000, 1.983, 9.400],
    [12.000, 1.983, 9.600],
    [1.000, 1.984, 9.600],
    [2.000, 1.984, 9.700],
    [3.000, 1.984, 9.800],
    [4.000, 1.984, 9.900],
    [5.000, 1.984, 9.900],
    [6.000, 1.984, 10.000],
    [7.000, 1.984, 10.100],
    [8.000, 1.984, 10.200],
    [9.000, 1.984, 10.300],
    [10.000, 1.984, 10.400],
    [11.000, 1.984, 10.400],
    [12.000, 1.984, 10.500]
])



# Transform coordinates into a cube shape
transformed_coordinates = transform_to_cube(coordinates2, desired_size=10.0)

# Print the transformed coordinates
for atom in transformed_coordinates:
    print(f"Fr {atom[0]:.3f} {atom[1]:.3f} {atom[2]:.3f}")
```
